## Advanced Python Skills

### 1. IF Statement

```python
import math

def get_log(x):
    if x > 0:
        y = math.log(x)
    else:
        y = float('nan')
    return y
```

```python
x = 5
log_val1 = get_log(x)

log_val2 = math.log(x) if x > 0 else float('nan') # advanced if in one sentence

print(log_val1)
print(log_val2)
```

### 2. For Loop

```python
# find the even numbers from 0 -- 1000
l1 = []
for i in range(1000):
    if i % 2 == 0:
        l1.append(i)
print(l1)
```

```python
l2 = [i for i in range(1000) if i % 2 == 0]
print(l2)
```

### 3. Data Structure in Python

- LIST

```python
# edit the element in the list
l = [1, 'a', 2, 'b']
l[0] = 3
l
```

```python
# add element in the last 
l.append(6)
l
```

```python
# eneraurte -- for loop
for i in l:
    print(i)
```

```python
# enerate -- while loop
i = 0
while i != len(l):
    print(l[i])
    i += 1
```

```python
# contact lists
[1, 2] + [3, 4]
```

```python
# repeat list
print(l * 5)
```

```python
# in the list or not
print(1 in [1, 2])
print(8 in l)
```
